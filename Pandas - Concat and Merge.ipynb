```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

## Example No 1

## Concat

```python
left = pd.DataFrame({'id':[1,2,3,4,5],
                     'Name': ['Shanu', 'Amar', 'Santosh', 'Rahul', 'Jayesh'],
                     'Lang':['English','Hindi','Marathi','Telugu','Gujarathi']})
right = pd.DataFrame({'id':[1,2,3,4,5],
                      'Name': ['John', 'Dinesh', 'Ranbir', 'Rahul', 'Jayesh'],
                      'Lang':['English','Hindi','Tamil','Bangali','Malyalam']})

pd.concat([left,right])
```

## Append

```python
left = pd.DataFrame({'id':[1,2,3,4,5],
                     'Name': ['Shanu', 'Amar', 'Santosh', 'Rahul', 'Jayesh'],
                     'Lang':['English','Hindi','Marathi','Telugu','Gujarathi']})
right = pd.DataFrame({'id':[1,2,3,4,5],
                      'Name': ['John', 'Dinesh', 'Ranbir', 'Rahul', 'Jayesh'],
                      'Lang':['English','Hindi','Tamil','Bangali','Malyalam']})

left.append([right])
```

## Merge

```python
left = pd.DataFrame({'id':[1,2,3,4,5],
                     'Name': ['Shanu', 'Amar', 'Santosh', 'Rahul', 'Jayesh'],
                     'Lang':['English','Hindi','Marathi','Telugu','Gujarathi']})
right = pd.DataFrame({'id':[1,2,3,4,5],
                      'Name': ['John', 'Dinesh', 'Ranbir', 'Rahul', 'Jayesh'],
                      'Lang':['English','Hindi','Tamil','Bangali','Malyalam']})

pd.merge(left,right,on='id')
```

## Left Merge

```python
left = pd.DataFrame({'id':[1,2,3,4,5],
                     'Name': ['Shanu', 'Amar', 'Santosh', 'Rahul', 'Jayesh'],
                     'Lang':['English','Hindi','Marathi','Telugu','Gujarathi']})
right = pd.DataFrame({'id':[1,2,3,4,5],
                      'Name': ['John', 'Dinesh', 'Ranbir', 'Rahul', 'Jayesh'],
                      'Lang':['English','Hindi','Tamil','Bangali','Malyalam']})

pd.merge(left, right, on='Lang', how='left')
```

## Right Merge

```python
left = pd.DataFrame({'id':[1,2,3,4,5],
                     'Name': ['Shanu', 'Amar', 'Santosh', 'Rahul', 'Jayesh'],
                     'Lang':['English','Hindi','Marathi','Telugu','Gujarathi']})
right = pd.DataFrame({'id':[1,2,3,4,5],
                      'Name': ['John', 'Dinesh', 'Ranbir', 'Rahul', 'Jayesh'],
                      'Lang':['English','Hindi','Tamil','Bangali','Malyalam']})

pd.merge(left, right, on='Lang', how='right')
```

## Inner Merge

```python
left = pd.DataFrame({'id':[1,2,3,4,5],
                     'Name': ['Shanu', 'Amar', 'Santosh', 'Rahul', 'Jayesh'],
                     'Lang':['English','Hindi','Marathi','Telugu','Gujarathi']})
right = pd.DataFrame({'id':[1,2,3,4,5],
                      'Name': ['John', 'Dinesh', 'Ranbir', 'Rahul', 'Jayesh'],
                      'Lang':['English','Hindi','Tamil','Bangali','Malyalam']})

pd.merge(left, right, on='Lang', how='inner')
```

## Outer Merge

```python
left = pd.DataFrame({'id':[1,2,3,4,5],
                     'Name': ['Shanu', 'Amar', 'Santosh', 'Rahul', 'Jayesh'],
                     'Lang':['English','Hindi','Marathi','Telugu','Gujarathi']})
right = pd.DataFrame({'id':[1,2,3,4,5],
                      'Name': ['John', 'Dinesh', 'Ranbir', 'Rahul', 'Jayesh'],
                      'Lang':['English','Hindi','Tamil','Bangali','Malyalam']})

pd.merge(left, right, on='Lang', how='outer')
```

```python
st_19 = pd.read_csv('D:\Data Science\Attachement\Sales Transactions-2019.csv')
st_19
```

```python
sales_concat_data = pd.concat([st_18,st_19])
sales_concat_data
```

```python
sales_append_data = st_18.append([st_19])
sales_append_data
```

## Left Merge

```python
sales_left_merge = pd.merge(st_18,st_19, on='Product', how='left')
sales_left_merge
```

## Right Merge

```python
sales_right_merge = pd.merge(st_18,st_19, on='Product', how='right')
sales_right_merge
```

## Outer Merge

```python
sales_outer_merge = pd.merge(st_18,st_19, on='Product', how='outer')
sales_outer_merge
```

## Inner Merge 

```python
sales_inner_merge = pd.merge(st_18,st_19, on='Product', how='inner')
sales_inner_merge
```
