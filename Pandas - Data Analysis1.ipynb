```python
football_stats = pd.read_csv('D:\Data Science\Attachement\Euro_2012_stats_TEAM.csv')
football_stats
```

### 2.1 Display first 5 observations of first 5 columns

### 2.3 Take all observations of Team, Shooting Accuracy, Red Cards and Yellow Cards.

```python
filtered_data = football_stats[['Team','Shooting Accuracy','Red Cards','Yellow Cards']]
filtered_data
```

### 2.4 Display the team which got more Number of Yellow Cards

```python
filtered_data.sort_values(by= 'Yellow Cards',ascending=False,inplace=True)
```

```python
filtered_data
```

### 2.5 Display the teams that Got Yellow Cards more than 7 and Red Cards 1.

```python
filtered_data[filtered_data['Yellow Cards']>7]
```

```python
filtered_data[filtered_data['Red Cards'] == 1]
```

```python
filtered_data[(filtered_data['Yellow Cards'] > 7) | (filtered_data['Red Cards'] == 1)]
```
